```python
import random

dice_drawing = {
    1: (
        "-----------",
        "|             |",
        "|      o     |",
        "|             |",
        "-----------",
    ),
    2: (
        "-----------",
        "|      o     |",
        "|             |",
        "|      o     |",
        "-----------",
    ),
    3: (
        "-----------",
        "|     o      |",
        "|     o      |",
        "|     o      |",
        "-----------",
    ),
    4: (
        "-----------",
        "|   o   o   |",
        "|            |",
        "|   o   o   |",
        "-----------",
    ),
    5: (
        "-----------",
        "|   o   o   |",
        "|     o      |",
        "|   o   o   |",
        "-----------",
    ),
    6: (
        "-----------",
        "|   o   o   |",
        "|   o   o   |",
        "|   o   o   |",
        "-----------",
    ),

}

def roll_dice():
  d1 = random.randint(1, 6)
  d2 = random.randint(1, 6)
  print(f"Dice rolled: {d1} and {d2} \n")
  print("\n".join(dice_drawing[d1]))
  print("\n".join(dice_drawing[d2]), '\n')

def ask_user():
  while True:
    res = input("Would you like to roll the dice? Enter 'Y' for yes, or 'N' for no: ")
    if res == "Y" or res == "y":
      roll_dice()
    elif res == "N" or res == "n":
      print("Play ended")
      break
    else: 
      print("Invalid input, please enter 'Y' for yes, or 'N' for no.  \n")     

ask_user() 
```
